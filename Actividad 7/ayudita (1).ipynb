### Árboles de decisión

```python
#Arbol de Regresion con una variable
X = df[['rm']]
y = df[["medv"]]
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.3, random_state = 10)
#Regresion
tree_regressor = DecisionTreeRegressor(max_depth= 2, random_state= 10)
#Etrenamiento
tree_regressor.fit(X_train , y_train)
y_pred = tree_regressor.predict(X_test)
print(f" R2  puntaje árbol de Regresión simple: {r2_score(y_test,y_pred)}")
print(f" MAE puntaje árbol de Regresión simple: {mean_absolute_error(y_test,y_pred)}")
print(f" MSE puntaje árbol de Regresión simple: {mean_squared_error(y_test,y_pred)}")
```

###  Random forest

```python
#Random forest simple 
X = df[["rm"]]
y = df[["medv"]]
X_train, X_test, y_train, y_test = train_test_split(X , y, test_size = 0.3, random_state = 10)
#Regresion
rfr = RandomForestRegressor()
#Etrenamiento
rfr.fit(X_train , y_train.values.ravel())
y_pred = rfr.predict(X_test)
print(f" R2  score RandomForestRegressor simple: {r2_score(y_test,y_pred)}")
print(f" MAE score RandomForestRegressor simple: {mean_absolute_error(y_test,y_pred)}")
print(f" MSE score RandomForestRegressor simple: {mean_squared_error(y_test,y_pred)}")
```
